```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/danielvieira95/Inteligencia_Artificial_2025/refs/heads/main/bases_de_dados/monitoramento_tempo.csv')
```

```python
import datetime
df['data'] = pd.to_datetime(df['data'])
```

```python
# A e C)
fig = plt.figure(figsize=(15,5)) # Tamanho da figura
eixo = fig.add_axes([0,0,1,1]) # Eixos do grfico
eixo.plot(df['data'],df['humidade'],color='b') # Dados no grfico
eixo.grid(True) # Vai ter grades
eixo.set_title('Umidade no momento',fontsize=25,pad=20) # Titulo do grfico
eixo.set_ylabel('Umidade',fontsize=20) # legenda eixo Y
eixo.set_xlabel('Ano',fontsize=20) # legenda eixo X
eixo.legend(['umidade'],loc='best') # legenda dos dados
```

```python
# B)
fig = plt.figure(figsize=(15,5)) # Tamanho da figura
eixo = fig.add_axes([0,0,1,1]) # Eixos do grfico
eixo.plot(df['data'],df['velocidade do vento'],color='b') # Dados no grfico
eixo.grid(True) # Vai ter grades
eixo.set_title('Velocidade do vento no momento',fontsize=25,pad=20) # Titulo do grfico
eixo.set_ylabel('Velocidade do vento',fontsize=20) # legenda eixo Y
eixo.set_xlabel('Ano',fontsize=20) # legenda eixo X
eixo.legend(['velocidade do vento'],loc='best') # legenda dos dados
```

```python
# D)
fig = plt.figure(figsize=(15,5)) # Tamanho da figura
eixo = fig.add_axes([0,0,1,1]) # Eixos do grfico
eixo.plot(df['data'],df['direo do vento'],color='b') # Dados no grfico
eixo.grid(True) # Vai ter grades
eixo.set_title('Direo do vento',fontsize=25,pad=20) # Titulo do grfico
eixo.set_ylabel('Direo do vento',fontsize=20) # legenda eixo Y
eixo.set_xlabel('Ano',fontsize=20) # legenda eixo X
eixo.legend(['direo do vento'],loc='best') # legenda dos dados
```

```python
# E)
fig = plt.figure(figsize=(15,5)) # Tamanho da figura
eixo = fig.add_axes([0,0,1,1]) # Eixos do grfico
eixo.plot(df['data'],df['presso'],color='b') # Dados no grfico
eixo.grid(True) # Vai ter grades
eixo.set_title('Presso',fontsize=25,pad=20) # Titulo do grfico
eixo.set_ylabel('Presso',fontsize=20) # legenda eixo Y
eixo.set_xlabel('Ano',fontsize=20) # legenda eixo X
eixo.legend(['presso'],loc='best') # legenda dos dados
```
